```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Load the data in the (yellow highlighted) Data Folder in the file auto   -mpg.data into a pandas DataFrame.

```python
autodata = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data', sep='\s+',header=None)
autodata
```

Rename data columns

```python
autodata.rename(columns={0:'mpg', 1:'cylinders', 2:'displacement', 3:'horsepower', 4:'weight', 5:'acceleration', 6:'model year', 7:'origin', 8:'car name'}, inplace=True)
autodata.columns
```

```python
autodata
```

Replace this â€˜?â€™ with a pandas-appropriate missing value

```python
autodata.replace('?', np.NaN)
```

Convert horsepower column to numeric

```python
autodata.horsepower = pd.to_numeric(autodata.horsepower, errors = 'coerce')
```

 Convert origin column values respectively to â€˜USAâ€™, â€˜Asiaâ€™, and â€˜Europeâ€™

```python
origin = {1: 'USA', 2: 'Asia',3: 'Europe'}

autodata.origin = autodata.origin.map(origin)
```

 Scatterplot that shows the relationship between horsepower and weight

```python
sns.lmplot("weight", "horsepower", data=autodata, hue='origin', fit_reg=True)
```

 Scatterplot that shows the relationship between acceleration and weight

```python
sns.lmplot("weight", "acceleration", data=autodata, hue='origin', fit_reg=True)
```

It seems that acceleration is affected by the weight in both USA and Europe. As the weight increases, the acceleration decreases. For Asia, the acceleration seems to stay constant regardless of weight.
