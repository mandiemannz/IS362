Question 1- What is the northernmost airport in the United States?
First thing we do is to read the data in panads. 


```python
airports = pd.read_csv('https://raw.githubusercontent.com/hadley/nycflights13/master/data-raw/airports.csv')
airports.head()
```

To figure out the northern most airport, we have to look at the data with the most latitiude. Latitudes of +90 and -90 degrees correspond to the north and south geographic poles on the earth, respectively

```python
airports.sort_values('lat', ascending=False).head()
```

Answer - Based on the latitude, the northern most airport is Wiley Post Will Rogers Memorial. The tzone of Dillant Hopkins Airport is unknown, so we cannot determine if it is in America or not based on the data that is given.

Question 3 - On February 12th, 2013, which New York area airport had the windiest weather?

```python
weather = pd.read_csv('https://raw.githubusercontent.com/hadley/nycflights13/master/data-raw/weather.csv')
weather.head()
```

```python
speediest = weather[(weather.month == 2) & (weather.day == 12) & (weather.year == 2013)]
speediest.sort_values('wind_speed', ascending = False).head()
```

The airport with the windiest weather is EWR on Feburary 12th, 2013. The wind speed was 1048.36058. This was done by first gathering all the wind speed data on Feburary 12th then sorting the values based on wind speed.
